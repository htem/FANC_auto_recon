```python
#from joblib import Parallel, delayed

#PROJECT PROPERTIES
deknife_channel_name = 'vnc1-deknife'
catmaid_project_voxel_size = (4.3, 4.3, 45)
img_root = "/mnt/gandalf_ssd_v2/vnc1_alignmentV3_full/"



#TUNABLE PARAMETERS
highest_zoom_level = 1 #The script always deknifes level 0 tiles. Set this to 1 or higher to also deknife higher levels.
z_range = 12 #How many sections to deknife both above and below the linked location (so in total 2*z_range+1 sections will be deknifed)
clahe = cv2.createCLAHE(clipLimit=1.6, tileGridSize=(2, 2)) #Increase clipLimit for more aggressive contrast adjustment, decrease for more conservative

#PROBABLY DON'T CHANGE THESE
apikey = REDACTED #htem slack bot
sc = SlackClient(apikey)
staged_directory = "staged"
backups_directory = "tile_backups_before_deknifemarking"

def find_id(chnnl):
    channels = sc.api_call("channels.list")
    channels = channels['channels']
    for i in range(0, len(channels)):
        z = channels[i]
        if z['name'] == chnnl:
            break
    return channels[i]['id']

def send_message(chnnl, message):
    channel_id = find_id(chnnl)
    sc.api_call("chat.postMessage", channel=channel_id, text=message)
    return chanel_id

def find_urls(msgs):
    ts = []
    to_run = []
    restore = []
    for i in range(0, len(msgs)):
        if 'replies' not in msgs[i].keys():
	    txt = msgs[i]['text']
	    t = msgs[i]['ts']
	    thisRestore = False
	    if txt[:8] == 'restore ':
		thisRestore = True
		txt=txt[8:]
	    if txt[:5] == '<http':
		to_run.append(txt)
		ts.append(t)
		restore.append(thisRestore)
	    #if 'replies' in msgs[i].keys():
		#return to_run, ts
    return to_run, ts, restore

def get_messages(chnnl):
    channel_id = find_id(chnnl)
    convos = sc.api_call("channels.history", channel=channel_id)
    return convos['messages']

def dftuv(M):
    u=np.arange(0, M)
    u[u>M/2]=u[u>M/2]-M
    return np.meshgrid(u, u)
# Butter
def lpfilter(t, M, D0, n=4):
    X, Y = dftuv(M)
    D=np.sqrt(X*X+Y*Y)
    if t == 'g':
        return np.exp(-np.power(D, 2)/(2*(D0**2)))
    elif t=='b':
        return 1./np.power(1 + D/D0, 2*n)
    elif t=='s':
        return (D < D0/np.sqrt(2)).astype(np.int)

def notch(t, M, D0, x, y, n=4):
    hlp = lpfilter(t, M, D0, n)
    H = 1 - hlp
    H= np.roll(H, y-1, axis=0)

    H= np.roll(H, x-1, axis=1)
    return H

def move(M, x, y):
    M= np.roll(M, y-1, axis=0)
    M= np.roll(M, x-1, axis=1)
    return M

def deknife_img(fn):
    zoom = int(fn.split('/')[-4])
    #print zoom
    im_size = 1024
    ratio = im_size/2048.*(2**zoom)
    H = notch('g', im_size, int(6*ratio), 0, 0, int(4*ratio)) # or s
    H_inv = 1.0 - H
    thresh = int(30*ratio)
    img=cv2.imread(os.path.join(fn), 0)
    # only do anything if image is not all black
    if img.sum() > 0:
        f = np.fft.fft2(img)
        fshift = np.fft.fftshift(f)
        mag_spec = 20*np.log(np.abs(fshift))
        rad = int(20 * ratio)
        E_arr = []
        for i in range(-15-77, 15-77):
            x = int(np.cos(np.deg2rad(i))*rad)
            y = int(np.sin(np.deg2rad(i))*rad)
            H_tmp = move(H_inv, x+im_size/2, y+im_size/2)+move(H_inv, im_size/2-x, im_size/2-y)
            E = (mag_spec*H_tmp).sum()
            E_arr.append((i, E))
        E_arr.sort(key=lambda x: x[1])
        deg = E_arr[-1][0]
        x = int(np.cos(np.deg2rad(deg))*rad)
        y = int(np.sin(np.deg2rad(deg))*rad)
        mult = move(H, x+im_size/2, y+im_size/2) * move(H, im_size/2-x, im_size/2-y)
        mult[im_size/2-int(ratio*2):im_size/2+int(ratio*2), im_size/2-int(ratio*2):im_size/2+int(ratio*2)] = 1.0
        fs_new = fshift *mult
        i_new_g = np.real(np.fft.ifft2(np.fft.ifftshift(fs_new)))
        print E_arr[-1][1]/E_arr[0][1]
        # we have to clip i_new_g to 0, 255. Any pixels above or below I set to 255
        i_new_g[ i_new_g > 255 ] = 255#img[ i_new_g > 255 ]
        i_new_g[ i_new_g < 0 ] = 0#img[ i_new_g < 0 ]

        mask = np.zeros(i_new_g.shape)
        w = int(64*ratio)
        d_img = img.astype(float) - i_new_g
        if d_img.max() == 0:
            im_partial = img
        else:
            for i in range(0, i_new_g.shape[0]-w+1, w):
                for j in range(0, i_new_g.shape[0]-w+1, w):
                    seg = d_img[i:i+w, j:j+w].astype(np.float)
                    mask[i:i+w, j:j+w] = seg.max() - seg.min()
            mask = mask > thresh
            im_partial = img.copy()
            im_partial[mask] = i_new_g[mask]
        im_partial = clahe.apply(im_partial)
        print "Writing to %s"%os.path.join(fn)

        cv2.imwrite(os.path.join(fn), im_partial)

def getPyramid(row, col, zoom, z):
    fns=[]
    fns.append(str(zoom)+'/'+str(z)+'/'+str(row)+'/'+str(col)+'.jpg')
    if zoom is not 0:
        fns.extend(getPyramid(row*2, col*2, zoom-1, z))
        fns.extend(getPyramid(row*2+1, col*2, zoom-1, z))
        fns.extend(getPyramid(row*2, col*2+1, zoom-1, z))
        fns.extend(getPyramid(row*2+1, col*2+1, zoom-1, z))
    return fns

def script():
    msgs = get_messages(deknife_channel_name)
    urls, ts, restore = find_urls(msgs)
    print urls, ts, restore
    for index in range(0, len(urls)):
        url = urls[index]
        t = ts[index]
        x = int(re.findall("xp=\d+", url)[0][3:])/catmaid_project_voxel_size[0]
        y = int(re.findall("yp=\d+", url)[0][3:])/catmaid_project_voxel_size[1]
        z = int(re.findall("zp=\d+", url)[0][3:])/catmaid_project_voxel_size[2]
        print x*catmaid_project_voxel_size[0], y*catmaid_project_voxel_size[1], z*catmaid_project_voxel_size[2]
        row=int(y/1024./(2**highest_zoom_level)-.5)
        col=int(x/1024./(2**highest_zoom_level)-.5)
        relativefns = []
        print "calculating files to operate on"
        for iz in range(z-z_range, z+z_range+1): #Loop over z slices and get filenames for each
            relativefns.extend(getPyramid(row, col, highest_zoom_level, iz))
            relativefns.extend(getPyramid(row+1, col, highest_zoom_level, iz))
            relativefns.extend(getPyramid(row, col+1, highest_zoom_level, iz))
            relativefns.extend(getPyramid(row+1, col+1, highest_zoom_level, iz))

        absolutefns=[]
        if restore[index]: #Restore from backups
            for relativefn in relativefns:
		absolutefn=os.path.join(img_root, staged_directory, relativefn)
		backupfn = os.path.join(img_root, backups_directory, relativefn)
		if not os.path.exists(backupfn):
		    continue
		if not os.path.exists(os.path.dirname(absolutefn)):
		    os.makedirs(os.path.dirname(absolutefn))
                #print "move("+backupfn+", "+absolutefn+")"
                shutil.move(backupfn, absolutefn)

        else: #Run deknifing and CLAHE
	    for relativefn in relativefns:
		absolutefn=os.path.join(img_root, staged_directory, relativefn)
		if not os.path.exists(absolutefn):
		    continue
		backupfn = os.path.join(img_root, backups_directory, relativefn)
		if os.path.exists(backupfn): #Skip running if a backup already exists
		    continue
		if not os.path.exists(os.path.dirname(backupfn)):
		    os.makedirs(os.path.dirname(backupfn))
		if not os.path.exists(backupfn):
		    shutil.copyfile(absolutefn, backupfn)
		absolutefns.append(absolutefn)
	    print "running"
	    pool = multiprocessing.Pool(25)
	    print(absolutefns)
	    pool.map(deknife_img, absolutefns)
            pool.terminate()
            del pool
        print "Posting to slack"
        channel_id = find_id(deknife_channel_name)
        if restore[index]:
            sc.api_call("chat.postMessage", channel=channel_id, text="restoring :done:", thread_ts = t)
        else:
            sc.api_call("chat.postMessage", channel=channel_id, text="deknifing and CLAHE :done:", thread_ts = t)
        print "Posted to slack"

if __name__ == '__main__':
    while True:
        script()
        print 'I schleep'
        time.sleep(10)
```
